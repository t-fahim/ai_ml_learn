Steamming

```python
def simple_stem(word):
    suffixes = ['ing', 'ed', 'ly', 'es', 's', 'ment', 'ness']
    for suffix in suffixes:
        if word.endswith(suffix):
            return word[:-len(suffix)]
    return word

# Example usage
words = ['running', 'studied', 'happiness', 'cares']
stems = [simple_stem(word) for word in words]
print(stems)  # Output: ['runn', 'studi', 'happi', 'care']
```

Lemmatization

```python
def simple_lemmatize(word):
    irregulars = {
        'better': 'good',
        'worse': 'bad',
        'caring': 'care',
        'studies': 'study',
        'running': 'run'
    }
    
    if word in irregulars:
        return irregulars[word]
        
    suffixes = {
        'ing': '', 'ed': '', 'ly': '', 'es': '', 's': '', 'ment': '', 'ness': ''
    }
    
    for suffix in suffixes:
        if word.endswith(suffix):
            return word[:-len(suffix)] + suffixes[suffix]
            
    return word
# Example usage
words = ['running', 'studies', 'better', 'caring']
lemmas = [simple_lemmatize(word) for word in words]
print(lemmas)  # Output: ['run', 'study', 'good', 'care']
```

Stemming using NLTK

```python
import nltk
from nltk.stem import PorterStemmer

# Downloading required NLTK resources
nltk.download('punkt')

# Example words
words = ["ladies", "studies", "happiness", "mapping", "better"]

# Initialize stemmer
porter = PorterStemmer()

# Apply stemming
porter_stems = [porter.stem(word) for word in words]

# Display results
print("Original Words:", words)
print("Porter Stemming:", porter_stems)
```

Lammatization using NLTk

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import nltk

# Downloading required NLTK resources
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to get POS tag for lemmatization
def get_pos(word):
    pos_tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}
    return tag_dict.get(pos_tag, wordnet.NOUN)

# Example words
words = ["running", "studies", "better", "caring", "happiness"]
# Apply lemmatization
lemmas = [lemmatizer.lemmatize(word, get_pos(word)) for word in words]

# Display results
print("Original Words:", words)
print("Lemmatization:", lemmas)
```
