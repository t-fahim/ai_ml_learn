```python
# load the dataset
# Importing libraries
import nltk
import numpy as np
import pandas as pd
import random
from sklearn.model_selection import train_test_split
import pprint, time

#download the treebank corpus from nltk
nltk.download('treebank')

#download the universal tagset from nltk
nltk.download('universal_tagset')

# reading the Treebank tagged sentences
nltk_data = list(nltk.corpus.treebank.tagged_sents(tagset='universal'))

#print the first two sentences along with tags
print(nltk_data[:2])
```

Printing some sample data

```python
# print each word along with its tag
for sent in nltk_data[:2]:
    for tuple in sent:
        print(tuple)
```

vectorization & split into test and train data

```python
# function for transforming documents into counts
from sklearn.feature_extraction.text import CountVectorizer
# function for encoding categories
from sklearn.preprocessing import LabelEncoder

# store the feature matrix (X) and response vector (y)
X = [tuple[0] for sent in nltk_data for tuple in sent]
y = [tuple[1] for sent in nltk_data for tuple in sent]

# pull the data into vectors
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(X)

encoder = LabelEncoder()
y = encoder.fit_transform(y)

# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)
```

Train and testing the mode

```python
# training the model on training set
from sklearn.naive_bayes import MultinomialNB

#initialize and train the MultinomialNB
nb = MultinomialNB()
nb.fit(X_train, y_train)

#testing the model
y_pred = nb.predict(X_test)

#convert test into original format
original_texts = vectorizer.inverse_transform(X_test)
# print(original_texts)
y_predicted_labels = encoder.inverse_transform(y_pred)
# print(y_predicted_labels)

## To print predicted data
# for text, label in zip(original_texts, y_predicted_labels):
#     print(f"Text: {' '.join(text)}")
#     print(f"Predicted Label: {label}")
#     print()

# comparing actual response values (y_test) with predicted response values (y_pred)
from sklearn import metrics
print("Multomial Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
print("Classification Report:\n", metrics.classification_report(y_test, y_pred))
```
