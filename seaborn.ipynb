# seaborn - data visualization library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
```

```python
# loading dataset from seaborn
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting theme
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x= 'total_bill', y= 'tip', col='time', hue='smoker', style='smoker', size='size')
```

```python
# load iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(data=iris, x= 'sepal_length', y= 'sepal_width', hue='species')
```

```python
# loading titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.countplot(data=titanic, x='class')
```

```python
sns.countplot(x='survived',data=titanic)
```

```python
sns.barplot(data=titanic, x='sex', y='survived', hue='class', )
```

```python
from sklearn.datasets import fetch_california_housing
house_california = fetch_california_housing()
house = pd.DataFrame(house_california.data, columns=house_california.feature_names)
house['price'] = house_california.target
house.head()
```

```python
sns.distplot(house['price'])
```

```python
# heat map
correlation = house.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8},cmap='Blues')
```
