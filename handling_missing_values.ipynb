```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading dataset to a pandas dataframe
dataset = pd.read_csv('/content/Placement_Dataset.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

Central tendencies:
- 1.mean
- 2.median
- 3.mode

```python
# analyse the distribution of data in salary column
sns.displot(data = dataset, x=dataset['salary']/1000, kde= True)
plt.xlabel("Salary (in thousands)")
plt.show()
```

Replace the missing value with median value

```python
median_value = dataset['salary'].median()
print(median_value)

# Fill the missing values with that number
dataset['salary'] = dataset['salary'].fillna(median_value)
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

```python
# analyse the distribution of data in salary column
sns.displot(data = dataset, x=dataset['salary']/1000, kde= True)
plt.xlabel("Salary (in thousands)")
plt.show()
```

Replacing the missing value with mean value

```python
mean_value = dataset['salary'].mean()
print(mean_value)

# Fill the missing values with that number
dataset['salary'] = dataset['salary'].fillna(mean_value)
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

```python
# analyse the distribution of data in salary column
sns.displot(data = dataset, x=dataset['salary']/1000, kde= True)
plt.xlabel("Salary (in thousands)")
plt.show()
```

Droping method

```python
salary_dataset = pd.read_csv('/content/Placement_Dataset.csv')
salary_dataset.head()
```

```python
salary_dataset.isnull().sum()
```

```python
# droping missing value
salary_dataset = salary_dataset.dropna(how='any')
```
