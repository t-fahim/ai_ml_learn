NumPy - Numerical Python
- Allow several mathematical operation including matrix
- Faster operation

```python
import numpy as np
```

```python
# campare python list and numpy list
from time import process_time
python_list = [i for i in range(10000)]
start_time = process_time()
python_list = [i+5 for i in python_list]
end_time = process_time()
total_time = end_time - start_time
print(total_time)
```

```python
from time import process_time
np_array = np.array([i for i in range(10000)])
start_time = process_time()
np_array += 5
end_time = process_time()
total_time = end_time - start_time
print(total_time)
```

# NumPy Arrays

```python
# list
list1 = np.array([1,2,3,4,5])
type(list1)
```

```python
# creating 1d array
a = np.array([1,2,3,4,5])
print(a)
a.shape
```

```python
# creating 2d array
b = np.array([(1,2,3,4),(5,6,7,8)])
print(b)
b.shape
```

```python
c = np.array([(1,2,3,4),(5,6,7,8)],dtype=float)
print(c)
c.shape
```

```python
# Creating a array of zeros
d = np.zeros((2,3))
print(d)
```

```python
# create a array of ones
e = np.ones((3,3))
print(e)
```

```python
# array of a perticular value
f = np.full((3,4),5.2)
print(f)
```

```python
# create a identity matrix
i = np.eye(3)
print(i)
```

```python
# creating a array of random values
x = np.random.random((3,4))
print(x)
```

```python
# random intiger value array within a specific range
y = np.random.randint(10,99,(3,4))
print(y)
```

```python
# array with evenly space values -- specifiying the number of value required
z = np.linspace(10,100,5)
print(z)
```

```python
# array with evenly space values -- specifing the step
z = np.arange(10,100,5)
print(z)
```

```python
# converting a list to numpy array
list1 = [1,2,4,5,0]
list2 = np.asarray((list1))
print(list2)
print(type(list1))
print(type(list2))
```

# Array Manipulation

```python
a = np.random.randint(1,10,(2,3))
```

```python
# transpose
trans = np.transpose(a)
print(trans)
```

```python
# reshaping a array
a = np.random.randint(1,10,(2,3))
b = a.reshape(3,2)
print(a)
print(b)
```
