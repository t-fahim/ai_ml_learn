# Label Encoding - converting the labels into numuric form

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

**label encoding for breast cancer dataset**

```python
cancer_data = pd.read_csv('/content/data.csv')
cancer_data.head()
```

```python
# finding the count for different label
cancer_data['diagnosis'].value_counts()
```

```python
# load the label encoder funciton
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data['diagnosis'])
```

```python
# appending the labels to the dataframe
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

- 0 --> Bening(B)
- 1 --> Malignant(M)




```python
cancer_data['target'].value_counts()
```

**Label encoding for iris data**

```python
iris_data = pd.read_csv('/content/iris_data.csv')
iris_data.head()
```

```python
iris_data['Species'].value_counts()
```

```python
# loading the label encoder
label_encode2 = LabelEncoder()
```

```python
iris_labels = label_encode2.fit_transform(iris_data['Species'])
```

```python
iris_data['labels'] = iris_labels
iris_data.head()
```
