# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin value')
plt.title('sin wave')
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cos value')
plt.title('cos wave')
plt.show()
```
